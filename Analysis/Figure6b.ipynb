```python
import pandas as pd
import matplotlib.pyplot as plt

files = [
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/CARD/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/Tangram/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/RCTD/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/scVI/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/Seurat/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/SingleR/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/spatialID/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/tacco/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/cell2location/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/DestVI/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/spatialDWLS/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/spotlight/Kidney_metric/metrics_results.csv',
]

legend_labels = [
    'CARD', 'Tangram', 'RCTD', 'scANVI', 'Seurat', 'SingleR',
    'Spatial-ID', 'TACCO', 'Cell2location', 'DestVI', 'SpatialDWLS', 'SPOTlight'
]

section_order = [
    'ShamL', 'ShamR', 'Hour4L', 'Hour4R', 'Hour12L', 'Hour12R',
    'Day2L', 'Day2R', 'Day14L', 'Day14R', 'Week6L', 'Week6R'
]

colors = {
    'RCTD': '#1f77b4',
    'Seurat': '#ff7f0e',
    'SingleR': '#1b9e77',
    'CARD': '#d62728',
    'Tangram': '#9467bd',
    'GraphST': '#c5b0d5',
    'Spatial-ID': '#8c564b',
    'TACCO': '#e377c2',
    'DestVI': '#7f7f7f',
    'scANVI': '#bcbd22',
    'Cell2location': '#9edae5',
    'SpatialDWLS': '#c49c94',
    'SPOTlight': '#A6D854',
}

metric_specs = [
    ('f1_macro', 'Macro F1'),
    ('accuracy', 'Accuracy'),
    ('precision', 'Precision'),
    ('recall', 'Recall'),
    ('ARI', 'ARI'),
]

data_by_method = {}
for file, method in zip(files, legend_labels):
    df = pd.read_csv(file)
    df['section'] = pd.Categorical(df['section'], categories=section_order, ordered=True)
    df = df.sort_values('section')
    data_by_method[method] = df

fig, axes = plt.subplots(2, 3, figsize=(24, 9), sharex=True)
axes = axes.ravel()

for idx, (metric_col, metric_title) in enumerate(metric_specs):
    ax = axes[idx]
    for method in legend_labels:
        df = data_by_method[method]
        if metric_col not in df.columns:
            continue
        ax.plot(
            df['section'],
            df[metric_col],
            marker='o',
            color=colors.get(method, '#000000'),
            linewidth=2,
            label=method,
        )

    ax.set_title(metric_title, fontsize=14)
    ax.set_xticks(range(len(section_order)))
    ax.set_xticklabels(section_order, rotation=45, ha='right', fontsize=9)
    ax.tick_params(axis='x', labelbottom=True)
    ax.grid(False)

axes[5].axis('off')

for i in [0, 3]:
    axes[i].set_ylabel('Score', fontsize=12)

handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, bbox_to_anchor=(0.985, 0.5), loc='center left', frameon=False)
plt.tight_layout(rect=[0.01, 0.02, 0.93, 0.99])
outpath = '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/metric/Kidney/Figure6b_5metrics_2x3.pdf'
plt.savefig(outpath, dpi=300, bbox_inches='tight')
plt.show()
print('Saved:', outpath)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

files = [
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/CARD/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/Tangram/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/RCTD/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/scVI/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/Seurat/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/SingleR/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/spatialID/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/tacco/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/cell2location/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/DestVI/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/spatialDWLS/Kidney_metric/metrics_results.csv',
    '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/spotlight/Kidney_metric/metrics_results.csv'
]

legend_labels = ['CARD', 'Tangram', 'RCTD', 'scANVI', 'Seurat', 'SingleR', 'Spatial-ID', 'TACCO',
                 'Cell2location', 'DestVI', 'SpatialDWLS', 'SPOTlight']

section_order = ['ShamL', 'ShamR', 'Hour4L', 'Hour4R', 'Hour12L', 'Hour12R',
                 'Day2L', 'Day2R', 'Day14L', 'Day14R', 'Week6L', 'Week6R']

metric = 'ARI'
colors = {
    'RCTD': '#1f77b4',
    'Seurat': '#ff7f0e',
    'SingleR': '#1b9e77',
    'CARD': '#d62728',
    'Tangram': '#9467bd',
    'GraphST': '#c5b0d5',
    'Spatial-ID': '#8c564b',
    'TACCO': '#e377c2',
    'DestVI': '#7f7f7f',
    'scANVI': '#bcbd22',
    'Cell2location': '#9edae5',
    'SpatialDWLS': '#c49c94',   
    'SPOTlight': '#A6D854'
}

plt.figure(figsize=(12, 6))
avg_results = []
for file, method in zip(files, legend_labels):
    df = pd.read_csv(file)

    df['section'] = pd.Categorical(df['section'], categories=section_order, ordered=True)
    df = df.sort_values('section')
    mean_ari = df[metric].mean()

    avg_results.append({
        "method": method,
        "mean_accuracy": mean_ari
    })

    color = colors.get(method, "#000000")

    plt.plot(
        df['section'], df[metric],
        marker='o',
        color=color,
        label=method,
        linewidth=2
    )
avg_df = pd.DataFrame(avg_results).sort_values(
    "mean_accuracy", ascending=False
)
print(avg_df)
plt.xlabel('Section', fontsize=14)
plt.ylabel(' ', fontsize=12)
plt.title('ARI', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.tight_layout()

outpath = '/maiziezhou_lab2/yuling/MouseSpinal/label_transfer/metric/Kidney/ARI_stage6_comparison_0413.pdf'
plt.savefig(outpath, dpi=300, bbox_inches='tight')
plt.show()
print("Saved:", outpath)
```
